```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('./data/musicdata_cleaned.csv')
# Convert song length to seconds for numeric analysis
def convert_length_to_seconds(length_str):
    try:
        mins, secs = map(int, length_str.split(':'))
        return mins * 60 + secs
    except:
        return 0
df['Length_sec'] = df['Length'].astype(str).apply(convert_length_to_seconds)
```

## Distribution of Song Length (in Seconds)

Visualizing how long typical Nepali songs are.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Length_sec'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Song Lengths')
plt.xlabel('Length (Seconds)')
plt.ylabel('Number of Songs')
plt.grid(True)
plt.show()
```

## Energy Levels Histogram

Shows how energetic the average Nepali song is.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Energy'], bins=25, color='green', edgecolor='black')
plt.title('Distribution of Song Energy Levels')
plt.xlabel('Energy')
plt.ylabel('Number of Songs')
plt.grid(True)
plt.show()
```

## Cumulative Distribution of Acousticness

Cumulative trend showing acoustic nature of songs.

```python
plt.figure(figsize=(10, 6))
sorted_acoustic = sorted(df['Acoustic'].dropna())
plt.plot(sorted_acoustic, range(len(sorted_acoustic)), color='purple')
plt.title('Cumulative Distribution of Acousticness')
plt.xlabel('Acoustic Score')
plt.ylabel('Cumulative Count')
plt.grid(True)
plt.show()
```
