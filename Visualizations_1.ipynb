```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances, manhattan_distances
from sklearn.neighbors import NearestNeighbors
df = pd.read_csv('./data/musicdata_cleaned.csv')
# Fix 'Length' column to seconds
df['Length'] = '0:' + df['Length'].astype(str)
df['Length'] = pd.to_timedelta(df['Length'], errors='coerce').dt.total_seconds()
df.dropna(inplace=True)
```

##  Joint Plot â€“ Valence vs Energy

```python
sns.jointplot(data=df, x='Valence', y='Energy', kind='kde')
plt.suptitle('Joint KDE Plot of Valence and Energy', y=1.02)
plt.show()
```

##  Bubble Chart â€“ Popularity Sized

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Valence'], df['Energy'], s=df['Pop.']*3, alpha=0.5, c='teal', edgecolors='w')
plt.xlabel('Valence')
plt.ylabel('Energy')
plt.title('Bubble Chart: Valence vs Energy sized by Popularity')
plt.grid(True)
plt.show()
```

## Distribution of All Audio Features

```python
features = ['BPM', 'Energy', 'Dance', 'Loud', 'Valence', 'Length', 'Acoustic', 'Pop.']
df[features].hist(bins=20, figsize=(15,10), color='steelblue', edgecolor='black')
plt.suptitle('Distribution of Audio Features')
plt.tight_layout()
plt.show()
```

##  Boxplots of Valence, Energy, Dance, Acoustic

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[['Valence', 'Energy', 'Dance', 'Acoustic']])
plt.title('Boxplots of Selected Features')
plt.grid(True)
plt.show()
```
